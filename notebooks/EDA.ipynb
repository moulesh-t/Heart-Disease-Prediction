```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../data/heart.csv')
```

```python
df.head()
```

```python
print("Data Types: ")
print(df.info())
```

```python
df.describe()
```

```python
print("Unique Values in the target column: ")
print(df['target'].value_counts())
```

```python
print("Unique Values per column: ")
print(df.nunique())
```

```python
plt.figure(figsize=(8,6))
df['target'].value_counts().plot(kind='bar')
plt.title('Target Distribution')
plt.savefig('../plots/target-distribution.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
# Correlation Plot
corr = df.corr()
plt.figure(figsize=(14,6))
sns.heatmap(corr, annot=True, cmap='plasma')
plt.title('Correlation Heatmap')
plt.savefig('../plots/correlation.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
# Pairplots of various features
sns.pairplot(df, hue='target', palette='plasma', diag_kind='kde')
plt.suptitle('Pair Plot Colored by Target', y=1.02)
plt.savefig('../plots/pairplot.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.boxplot()
plt.title('Box Plots for Outlier Detection')
plt.xticks(rotation=45)
plt.savefig('../plots/boxplots_outliers.png', dpi=300, bbox_inches='tight')
plt.show()
```
